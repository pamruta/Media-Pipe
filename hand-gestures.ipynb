```python
import cv2
```

```python
# import mediapipe library

import mediapipe as mp
mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
```

```python
# load sample image
from google.colab.patches import cv2_imshow
image_file = "sample.jpg"
image = cv2.imread(image_file)
cv2_imshow(image)
```

```python
# convert to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# run pose detection
hands = mp_hands.Hands(max_num_hands=2)
output = hands.process(rgb_image)
```

```python
# display pose landmark points
if output.multi_hand_landmarks:
  for landmarks in output.multi_hand_landmarks:
    mp_draw.draw_landmarks(image, landmarks, mp_hands.HAND_CONNECTIONS)
    cv2_imshow(image)
else:
  print("Couldn't detect any human body..")
```
