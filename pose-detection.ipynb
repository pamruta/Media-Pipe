```python
import cv2
```

```python
# import mediapipe library

import mediapipe as mp
mp_pose = mp.solutions.pose
mp_draw = mp.solutions.drawing_utils
```

```python
# load sample image
from google.colab.patches import cv2_imshow
image_file = "sample.jpg"
image = cv2.imread(image_file)
cv2_imshow(image)
```

```python
# convert to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# run pose detection
pose = mp_pose.Pose()
output = pose.process(rgb_image)
```

```python
# display pose landmark points
if output.pose_landmarks:
  mp_draw.draw_landmarks(image, output.pose_landmarks, mp_pose.POSE_CONNECTIONS)
  cv2_imshow(image)
else:
  print("Couldn't detect any human body..")
```

```python
# get specific landmark point
print(output.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW])
```
