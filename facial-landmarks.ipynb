```python
import cv2
```

```python
# import mediapipe library

import mediapipe as mp
mp_facemesh = mp.solutions.face_mesh
mp_draw = mp.solutions.drawing_utils
```

```python
# load sample image
from google.colab.patches import cv2_imshow
image_file = "sample.jpg"
image = cv2.imread(image_file)
cv2_imshow(image)
```

```python
# convert to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# run face mesh model
face = mp_facemesh.FaceMesh(static_image_mode=True, max_num_faces=1, refine_landmarks=True)
output = face.process(rgb_image)
```

```python
# display facial landmark points
if output.multi_face_landmarks:
  for landmarks in output.multi_face_landmarks:
    mp_draw.draw_landmarks(image, landmarks, mp_facemesh.FACEMESH_CONTOURS)
    cv2_imshow(image)
else:
  print("Couldn't detect any face..")
```
